```python
import pandas as pd
import numpy as np

data1 = np.random.rand(1000000, 10)
df1 = pd.DataFrame(data1)

data2 = np.random.rand(1000000, 10)
df2 = pd.DataFrame(data2)
```

02.

```python

df_add = df1.add(df2)

df_mul = df1.mul(df2)

df_sub = df1.sub(df2)

print("Addition:")
print(df_add.head())

print("\nMultiplication:")
print(df_mul.head())

print("\nSubtraction:")
print(df_sub.head())
```

03.

```python
%timeit df_add = df1.add(df2)
print("Addition:")
df_add = df1.add(df2)
print(df_add.head())

%timeit df_mul = df1.mul(df2)
print("\nMultiplication:")
df_mul = df1.mul(df2)
print(df_mul.head())

%timeit df_sub = df1.sub(df2)
print("\nSubtraction:")
df_sub = df1.sub(df2)
print(df_sub.head())
```

03.

```python

import pandas as pd
import numpy as np
import seaborn as sns

iris = sns.load_dataset('iris')

print(iris.head())
print(iris.info())
print(iris.describe())
print(iris.isnull().sum())
```

04.

```python

print(iris.dtypes)

iris['species'] = iris['species'].astype('category')

iris.dtypes
```

05.

```python
import matplotlib.pyplot as plt

for column in iris.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=iris[column])
    plt.title(f'Box Plot of {column}')
    plt.show()
```

06.

```python
import matplotlib.pyplot as plt

for column in iris.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(6, 4))
    plt.hist(iris[column], bins=10, edgecolor='black')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')
    plt.show()
```

07.

```python


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.scatter(iris['sepal_length'], iris['sepal_width'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs. Sepal Width')

plt.subplot(2, 2, 2)
plt.scatter(iris['petal_length'], iris['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs. Petal Width')

plt.subplot(2, 2, 3)
plt.scatter(iris['sepal_length'], iris['petal_length'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Sepal Length vs. Petal Length')

plt.subplot(2, 2, 4)
plt.scatter(iris['sepal_width'], iris['petal_width'])
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.title('Sepal Width vs. Petal Width')

plt.tight_layout()
plt.show()
```

08.

```python
sns.pairplot(iris, hue='species')
plt.show()
```

09.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

numerical_features = iris.select_dtypes(include=['number'])

plt.figure(figsize=(8, 6))
correlation_matrix = numerical_features.corr()  # Calculate correlation for numerical features only
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```
